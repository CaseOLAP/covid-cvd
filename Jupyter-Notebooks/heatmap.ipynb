### Heatmap

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json as json
import seaborn as sns
sns.set(font_scale = 0.80)
%matplotlib inline
```

--------------

```python
data = pd.read_csv('../result/caseolap.csv')
data = data.set_index('protein')
ndf = data.copy(deep =True)
ndf.head()
```

-------------

```python
hdata = ndf.copy(deep =True)
hdata.max()
```

```python
plt.figure(figsize = [10,18])
sns.heatmap(hdata,\
            cmap="Blues",\
            #cmap = sns.cubehelix_palette(1000),\
            #cmap = sns.cubehelix_palette(8, start=.5, rot=-.75),\
            #cmap = sns.color_palette("Blues"),\
            yticklabels=True,\
            #annot = True,\
            vmin = 0.0,vmax = 0.30)
plt.savefig('plots/heatmap.pdf')
plt.savefig('plots/heatmap.png')
```

--------------

```python
with open("data/id2name.json","r")as f:
    id2name = json.load(f)

names = []
for item in ndf.index:
    names.append(id2name[item])

ndf['names'] =names
ndf.head(1)
```
